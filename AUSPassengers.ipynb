```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in the file and just pull out 2 columns
df = pd.read_csv("./resources/data/AUS_monthly_passengers.csv")[['Month', 'Total Passengers']]

#convert the Month to a datetime value
df['Month'] = pd.to_datetime(df['Month'])

df1 = df.sort_values(by='Month').set_index('Month').reset_index()
```

```python
df1.plot(x="Month", y="Total Passengers", kind="line", grid=True)
plt.xlabel("By month, from June 2013 to October 2018")
plt.ylabel("Number of Passengers")
plt.title("Passenger Volume Growth at ABIA")
plt.savefig("./resources/images/monthly_passengers.png")
plt.show()
```
