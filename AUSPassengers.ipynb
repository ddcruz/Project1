```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = "AUS_passengers.csv"
```

```python
csv_reader = pd.read_csv(csv_file)
df = pd.DataFrame(csv_reader)
df.head()
```

```python
new_df = df[['Month', 'Total Passengers']]
new_df.head()
```

```python
new_df['Month'].astype(str)
```

```python
new_df['Year'] = ""
new_df['Months'] = ""
new_df.head()
```

```python
bydate = new_df['Month'].str.split("/", expand=True)
```

```python
new_df['Year'] = bydate[2]
new_df['Months'] = bydate[0]
```

```python
new_df.head()
```

```python
byyear = new_df['Year'].str.split(" ", expand=True)

new_df['Years'] = byyear[0]
new_df.head()
```

```python
final_df = new_df[['Months', 'Years', 'Total Passengers']]
final_df.head()
```

```python
df2 = final_df.sort_values(by=['Years', 'Months'])
df2.head()
```

```python
reset_df = df2.reset_index(drop=True)
reset_df.head()
```

```python
reset_df.plot(x="Months", y="Total Passengers", kind="line")
plt.xlabel("By month, from June 2013 to October 2018")
plt.ylabel("No. of Passengers")
plt.title("Passenger Volume Growth at ABIA")
plt.savefig("monthly_passengers")
```
