```python
import pandas as pd
import requests 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#def get_data(year=2007, airport_1='AUS', data_request_type=0):
    #load the data
df =  pd.read_csv('./resources/data/September2018Arrival.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Labels for the sections of our pie chart
labels = ['On-Time Arrivals', 'Late Arrivals', 'Cancelled', 'Diverted']

# The values of each section of the pie chart
sizes = [79.46, 18.65, 1.70, 0.20]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral", "lightgreen"]

# Build Pie Chart
plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=0)
```

```python
# plot percent on-time arrival over the years
df.plot(x='Year', y='Percent On-Time Arrivals')
```

```python
# gca stands for 'get current axis'
ax = plt.gca()

# plot 4 lines
df.plot(kind='line',x='Year',y='Percent On-Time Arrivals',ax=ax)
df.plot(kind='line',x='Year',y='Percent Late Arrivals', color='red', ax=ax)
df.plot(kind='line',x='Year',y='Percent Cancelled', color='green', ax=ax)
df.plot(kind='line',x='Year',y='Percent Diverted', color='yellow', ax=ax)

plt.show()
```
