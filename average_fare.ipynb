```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
csv_file = "AIBA_Average_Airfare_2007_2017.csv"

csv_reader = pd.read_csv(csv_file)

aus_airfare_df = pd.DataFrame(csv_reader)
aus_airfare_df
```

```python
aus_airfare_df.plot(x="Year", y="Average Fare", kind="line")
```

```python
aus_airfare_df.plot(x="Year", y=["Average Fare", "Inflation Adjusted Avg Fare"], kind="line", grid=True)
plt.title("Average Fare and Inflation-Adjusted Fare,\n Austin Bergstrom International Airport, 2007-2017")
plt.ylabel("Fare in U.S. Dollars")
```
