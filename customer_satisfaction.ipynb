```python
import pandas as pd
import requests
```

```python
base_url = "https://data.austintexas.gov/resource/y48g-27kc.json"
```

```python
results = requests.get(base_url).json()
```

```python
results
```

```python
df = pd.DataFrame(results)
```

```python
df.head()
```

```python
df.to_csv("customers.csv")
```

```python
df.dropna(axis=0, how='any', inplace=True)
df.head()
```

```python
df = pd.t
```

```python
df['Overall'] = df.mean(axis=1)
df.head()
```
