```python
import pandas as pd
import requests 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#def get_data(year=2007, airport_1='AUS', data_request_type=0):
    #load the data
df =  pd.read_csv('./resources/data/September2018Arrival.csv')
```

```python
df.head()
```

```python
df = df[['Year', 'Late Arrivals', 'Late Departures', 'Cancelled', 'Diverted']]
df
```

```python
df.dtypes
```

```python
df = df.astype(float)
```

```python
# gca stands for 'get current axis'
ax = plt.gca()

# plot 4 lines
df.plot(kind='line',x='Year',y='Late Arrivals',ax=ax)
df.plot(kind='line',x='Year',y='Late Departures', color='red', ax=ax)
df.plot(kind='line',x='Year',y='Cancelled', color='green', ax=ax)
df.plot(kind='line',x='Year',y='Diverted', color='yellow', ax=ax)

plt.show()
```
