```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
import csv
```

```python
#csv_path = "Resources/Airline-On-Time1.csv
On_Time_df= pd.read_csv("Airline-On-Time.csv", skiprows=2)
On_Time_df
```

```python
year_pct_df = On_Time_df[ ['Year', 'Percent On-Time Arrivals'] ]
```

```python
# 
import matplotlib
import matplotlib.ticker as ticker

fig, ax = plt.subplots(1,1)
ax.xaxis.set_major_locator(matplotlib.dates.AutoDateLocator())
ax.xaxis.set_minor_locator(matplotlib.dates.AutoDateLocator())
ax.plot(year_pct_df['Year'], year_pct_df['Percent On-Time Arrivals'], color="orange", label="Years")

plt.legend(loc="best")
plt.title("Arrivals On Time Through the years")
plt.xlabel("years")
plt.ylabel("Percent Arrivals On Time")
locs, labels = plt.xticks()

plt.setp(labels, rotation=90)
plt.show()
plt.tight_layout()
```
